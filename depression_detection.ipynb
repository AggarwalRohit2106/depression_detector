```python
def text_cleaner(data):
    data=data.lower()
    
        
    return data  
```

```python
dff['message']=dff['message'].apply(text_cleaner)
```

```python
dff.head()
```

```python
def word_token(data):
    data=nltk.word_tokenize(data)
    return data
```

```python
dff['message']=dff['message'].apply(word_token)
```

```python
dff.head()
```

```python
def spec(data):
    new_data=[]
    for i in data:
        if i.isalnum():
            new_data.append(i)
    data=new_data[:]
    new_data.clear()
    return data
```

```python
dff['message']=dff['message'].apply(spec)
```

```python
dff.head()
```

```python
dff['message'][10]
```

```python
from tqdm import tqdm
```

```python
def stopword(data):
    new_data=[]
    for i in tqdm(data):
        
        if i not in stopwords.words('english') and i not in string.punctuation:
            new_data.append(i)
    
    data=new_data[:]
    new_data.clear()
    return data
```

```python
dff['message']=dff['message'].apply(stopword)
```

```python
dff.head()
```

```python
def stem(data):
    new_data=[]
    for i in tqdm(data):
        
        new_data.append(ps.stem(i))
    return " ".join(new_data) 
```

```python
dff['message']=dff['message'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=3000,stop_words='english')
```

```python
X=cv.fit_transform(dff['message']).toarray()
```

```python
X.shape
```

```python
Y=dff['label'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,classification_report
```

#### Naive based algo

```python
gnb.fit(X_train,y_train)
pred1=gnb.predict(X_test)
print(confusion_matrix(y_test,pred1))
print(accuracy_score(y_test,pred1))
print(precision_score(y_test,pred1))
print(classification_report(y_test,pred1))
```

```python
mnb.fit(X_train,y_train)
pred2=mnb.predict(X_test)
print(confusion_matrix(y_test,pred2))
print(accuracy_score(y_test,pred2))
print(precision_score(y_test,pred2))
print(classification_report(y_test,pred2))
```

```python
bnb.fit(X_train,y_train)
pred3=bnb.predict(X_test)
print(confusion_matrix(y_test,pred3))
print(accuracy_score(y_test,pred3))
print(precision_score(y_test,pred3))
print(classification_report(y_test,pred3))
```
